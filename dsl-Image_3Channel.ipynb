```python
import pywt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

def convt2image(vals):

    newArr=vals
    newArr=np.array(newArr)
    while newArr.shape[0]!=3:
        this_arr=[]
        for i in range(newArr.shape[0]-1):
            val1=newArr[i][:50]
            val2=newArr[i+1][:50]
            avg=(val1+val2)/2
            this_arr.append(avg)
        this_arr=np.array(this_arr)
        newArr=this_arr
#         print(newArr.shape)

    newArr=(newArr-np.min(newArr))/(np.max(newArr)-np.min(newArr))    

    new_img=np.zeros((7,7,3))



    for arr in newArr:
        channel=[]
        for i in range(7):
            for j in range(7):
                ctr=i*7+j
                new_img[i][j][0]=newArr[0][ctr]
                new_img[i][j][1]=newArr[1][ctr]
                new_img[i][j][2]=newArr[2][ctr]            

    return new_img
```

```python
df=pd.read_csv("data/DSL-StrongPasswordData.csv")
```

```python
classes=list(df.subject.unique())
class_labels=[i for i in range(len(classes))]
df['subject'].replace(classes,
                        class_labels, inplace=True)
```

```python
wavelists=[]
for f in pywt.families():
    print(f)
    if "gaus" in f:
        break
    wave_list=pywt.wavelist(family=f, kind='discrete')
    for w in wave_list:
        print("\t",w)
        wavelists.append(w)
    
```

```python
subjects=df.subject.unique()
for subject in subjects:
    df_f=df[df["subject"]==subject]
    count=0
    for index,row in df_f.iterrows():

        subject=row["subject"]
        sessionIndex=row["sessionIndex"]
        rep=row["rep"]
        name="sub_"+str(subject)+"_sessn_"+str(sessionIndex)+"_rep_"+str(rep)
        X=row[3:].values
    #     print(X.shape)
        vals=[]
        for w in wavelists:
            cA,cD=pywt.dwt(X,w)
            if cA.shape[0]>50:
    #             print(w)            
    #             print(cA.shape,cD.shape)
                vals.append(cA)
                vals.append(cD) 
        img3d=convt2image(vals)
        plt.title(name)
        plt.imshow(img3d)
        plt.savefig("dsl-img-pyWtSuccAvg/"+name+".svg")
        plt.savefig("dsl-img-pyWtSuccAvg/"+name+".pdf")        
        plt.show()
        count+=1
        if count>2:
            break
```
